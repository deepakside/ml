```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

# (A)

```python
#(a) line plot
x = np.arange(0,2*np.pi,0.1) # start,stop,step
y = np.sin(x) # sin() function
z = np.cos(x) #cos() function
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 2pi') # X label
plt.ylabel('sin(x) and cos(x)') # Y Label
plt.title('Plot of sin and cos from 0 to 2pi') #adding title
plt.legend(['sin(x)', 'cos(x)']) # legends
plt.show()
```

# (B)

```python
#(b) bar plot
# creating the dataset
sub_data = {'Machine Learning':15, 'Artificial Intelligence':25, 'DataMining':12, 'Project':10,'Information security':13}
stu_subjects = list(sub_data.keys())
students = list(sub_data.values())
 
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(stu_subjects, students, color ='green',width = 0.7)
plt.xlabel("Subjects Offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different Subjects in 6 semester")
plt.show()
```

# (C)

```python
#(c) scatter plot
x = np.random.normal(1.0, 5.0, 1000)
y = np.random.normal(10.0, 20.0, 1000)
plt.scatter(x, y)
plt.show()
```

# (D)

```python
#(d) pie plot
plt.pie(students, labels = stu_subjects)
plt.legend(title = "Five_subjects",loc='upper right')
plt.show() 
```

# (E)

```python
#(e) area plot

# Create data
x=range(1,6)
y=[1,4,6,8,4]

# Area plot
plt.fill_between(x, y,color='yellow')
plt.show()
```

# (F)

```python
#(f) histogram
x = np.random.normal(170, 10, 250)
plt.hist(x,color='orange')
plt.show() 
```

```python
#Generate different subplots from a given plot and color plot data
# Data for plotting
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.cos(2 * np.pi * x)
# Creating 6 subplots and unpacking the output array
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)
ax1.plot(x, y, color="red")
ax2.plot(x, y, color="blue")
ax3.plot(x, y, color="magenta")
ax4.plot(x, y, color="green")
ax5.plot(x, y, color="brown")
ax6.plot(x, y, color="yellow")
```
